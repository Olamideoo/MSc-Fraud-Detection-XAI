```python
import pandas as pd

# Load the clean dataset from preprocessing
data = pd.read_csv("../Data/cleaned_fraud_dataset.csv")

# Create a copy that represents the dataset after EDA
data.to_csv("../Data/eda_completed_dataset.csv", index=False)

data.head()
```

```python
#identify the categorical and numerical feature 
#identify the target variable
data.info()
data.columns
```

```python
#new account indicator
data['new_account'] = (  data['account_age_days'] < 180).astype(int)
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

print("Numerical:", numerical_cols)
print("Categorical:", categorical_cols)
```

HIGH TRANSACTION AMOUNT 

```python
# Create feature
data['high_value_transaction'] = (
    data['transaction_amount'] >
    data['transaction_amount'].quantile(0.75)
).astype(int)

# Verify feature
print("High Value Transaction")
print(data['high_value_transaction'].value_counts())
```

```python
print(pd.crosstab(data['high_value_transaction'], data['fraud_flag']))
```

HIGH LOGIN ATTEMPT 

```python
# Create feature
data['high_login_attempts'] = (
    data['login_attempts'] >= 3
).astype(int)

# Verify feature
print("\nHigh Login Attempts")
print(data['high_login_attempts'].value_counts())
```

```python
data['high_login_attempts'] = (
    data['login_attempts'] > data['login_attempts'].quantile(0.75)
).astype(int)

print("High Login Attempts")
print(data['high_login_attempts'].value_counts())
```

```python
print(data['login_attempts'].describe())
print(data['login_attempts'].value_counts().sort_index())
```

High Velocity Transaction

```python
# Create feature
data['high_velocity'] = (
    data['transaction_velocity_score'] >
    data['transaction_velocity_score'].quantile(0.75)
).astype(int)

# Verify feature
print("\nHigh Velocity")
print(data['high_velocity'].value_counts())
```

NIGHT TRASNACTION

```python
# Create feature
data['night_transaction'] = (
    (data['transaction_time_hour'] >= 22) |
    (data['transaction_time_hour'] <= 5)
).astype(int)

# Verify feature
print("\nNight Transaction")
print(data['night_transaction'].value_counts())
```

Far Distance Transaction

This feature identifies transactions occurring at unusually large geographical distances (top 25% of distances). Transactions taking place far from a customer's typical location may indicate fraudulent activity.

```python
# Create feature
data['far_distance_transaction'] = (
    data['geo_distance_km'] >
    data['geo_distance_km'].quantile(0.75)).astype(int)

# Verify feature
print("Far Distance Transaction")
print(data['far_distance_transaction'].value_counts())
```

```python
print(pd.crosstab(
    data['far_distance_transaction'],
    data['fraud_flag']))
```

High Device Risk

 This feature flags transactions originating from devices with risk scores in the top 25% of the dataset. High-risk devices may have a greater likelihood of being associated with fraudulent transactions.

```python
# Create feature
data['high_device_risk'] = (
    data['device_risk_score'] >
    data['device_risk_score'].quantile(0.75)).astype(int)

# Verify feature
print("High Device Risk")
print(data['high_device_risk'].value_counts())
```

```python
#checking if the new features are correlated with the target variable
print(pd.crosstab(
    data['high_device_risk'],
    data['fraud_flag']))
```
